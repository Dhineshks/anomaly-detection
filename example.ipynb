# Density estimation of data using gausian distribution
* Dataset : $\{x^{(1)}, \dots, x^{(m)} \}$ 
* say $x^{(i)} \in \mathbb{R}^n$ 
* x is a distributed gaussian with mean $\mu$ and variance $\sigma^2$ 

## 1. Gaussian distribution formula  $$ p\left( x; \mu, \sigma^2 \right) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{\left(x-\mu\right)^2}{2\sigma^2}}$$

## 2. parameter estimation
### 2.1. mean  $$ \mu_i = \frac{1}{m} \sum_{j=1}^m x_i^{(j)}$$
### 2.2. variance  $$ \sigma_i^2 = \frac{1}{m} \sum_{j=1}^m \left( x_i^{(j)} - \mu_i \right)^2$$
