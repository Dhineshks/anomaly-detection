# Density estimation of data using gausian distribution
* Dataset : $\{x^{(1)}, \dots, x^{(m)} \}$ 
* say $x^{(i)} \in \mathbb{R}^n$ 
* x is a distributed gaussian with mean $\mu$ and variance $\sigma^2$ 

## 1. Gaussian distribution formula  $$ p\left( x; \mu, \sigma^2 \right) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{\left(x-\mu\right)^2}{2\sigma^2}}$$

## 2. parameter estimation
### 2.1. mean  $$ \mu_i = \frac{1}{m} \sum_{j=1}^m x_i^{(j)}$$
### 2.2. variance  $$ \sigma_i^2 = \frac{1}{m} \sum_{j=1}^m \left( x_i^{(j)} - \mu_i \right)^2$$

```python
mu,sigma_square = func.estimateGaussian(df)
p = func.GaussianDistribution(df,mu,sigma_square)
```


# Algotithm Evaluation
* On cross validation set, predict,
if  $p(x)\lt$ $\varepsilon$ then it is anomaly, if $p(x)\ge$ $\varepsilon$ then it is normal

## Possible Evaluation Metrics
* True Positive, False Positive, False Negative, True Negative
* Precision / Recall
* F1 Score

## 1. F1 Score 
$$ F_1 = \frac{2 \cdot prec \cdot rec}{prec + rec} $$

### 1.1. Precision
$$ prec = \frac{tp}{tp + fp} $$

### 1.2. recall
$$ rec = \frac{tp}{tp + fn} $$

## Note :
* TP is True Positives - It insists it's an anomaly and the algorithm correctly classified it as an anomaly
* FP is False Positives - It insists it's not an anomaly, but the algorithm incorrectly classified it as an anomaly
* FN is False Negatives - It insists it's an anomaly, but the algorithm incorrectly classified it as not anomaly
